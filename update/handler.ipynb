```python
# 数据
dates = ['2023-02-06', '2023-02-13', '2023-02-20', '2023-02-27']

data = {'120412.SZ': [[110.02,34.12], [114.02,43.12], [118.02,45.12], [108.02,38.12]],
        '114125.SH': [[115.02,36.15], [117.99,22.13], [135.02,39.52], [110.02,32.42]],
        '112799.SH': [[123.02,24.43], [127.95,26.18], [132.02,30.12], [139.02,22.48]], 
        '128901.SZ': [[135.02,25.01], [125.99,22.43], [138.02,27.36], [113.02,32.62]],
       }
# 题
# 给出时间数据，dates，给出可转债时间序列数据字典，key为可转债代码，value为时间序列数据
# 对于时间序列数据，条元素对应dates的时间，且每条元素格式为【收盘价，转股溢价率】
# 双低策略定义为。。。，选双低前二作为持仓！
# 求出双低策略下，每个日期的选股，以及四个交易日后的最终收益 （也可考每周收益）

# 我写一下

# 一行提取四个转债代码
codes = data.keys()
sum_profit_deng = 0
sum_profit_jia = 0
previous=[]
for date_no in range(len(dates)):
    # 当前交易日选股
    choose = []


    # 选出双低最低的
    
    # 设置一个高的双低值
    low_price_1 = 200
    low_code_1 = 'no1'
    
    for code in codes:
        re = data[code][date_no][0] + data[code][date_no][1]
        if re < low_price_1:
            low_price_1 = re
            low_code_1 = code
            
    # 选出双低第二低的
    # 设置一个高的双低值
    low_price_2 = 200
    low_code_2 = 'no2'
    for code in codes:
        re = data[code][date_no][0] + data[code][date_no][1]
        if code != low_code_1 and re < low_price_2:
            low_price_2 = re
            low_code_2 = code 
      
    # 这里就选好股了
    choose = [low_code_1,low_code_2]
    print('选股', dates[date_no], choose[0], round(low_price_1,2), choose[1], round(low_price_2,2))
    

    # 我这里写了两种做法，等权和加权，我觉得没必要逼人写出某一种，随便写，只要是等权和加权之一就行
    if date_no >= 1:
        # 加权，价格高权重大
        current_price = 0
        last_price = 0
        for code in previous:
            current_price += data[code][date_no][0]
            last_price += data[code][date_no-1][0]
            
        profit_jia = current_price/last_price - 1
        
        sum_profit_jia += profit_jia
        
        # 等权，每个权重一样
        profit_deng = 0
        
        for code in previous:
            profit_deng += (data[code][date_no][0]/data[code][date_no-1][0]-1)
            
        profit_deng /= 2
        
        sum_profit_deng += profit_deng
        
        print('收益统计', dates[date_no])
        print('等权持仓','单周收益', round(profit_deng*100,2),'%')
        print('加权持仓','单周收益', round(profit_jia*100,2),'%')
        print('等权持仓','累计收益', round(sum_profit_deng*100,2),'%')
        print('加权持仓','累计收益', round(sum_profit_jia*100,2),'%')        
        
        
        
    # 记录上个交易日选股
    previous = choose.copy()
```
